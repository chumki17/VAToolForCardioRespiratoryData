```python
# transform returns array of float elements
def transform(x):
  x = x[1:-1]
  to_ret = []
  for item in x.split(","):
    item = item.strip()
    to_ret.append(float(item))
  return to_ret
```

```python
# Training Intensity Reduction
```

```python
def training_intensity_reduction(person : int):
    duration_tmhr = 0
    duration_thr_hr = 0
    age = df['Age'][person]
    TMHR = 208 - 0.7*age
    thrHR = 0.85*TMHR
    HR = transform(df['Heart_Rate'][person])
    duration_ten_percent = 0.1*len(HR)
    for h_rate in HR:
        if h_rate > TMHR:
            duratYellowion_tmhr = duration_tmhr + 1
        
        if h_rate > thrHR:
            duration_thr_hr = duration_thr_hr + 1
            
    # print("Duration for TMHR: ", duration_tmhr)
    # print("Duration for thrHHR: ", duration_thr_hr)
    # print("10% of total duration: ", 0.1*len(HR)) 
    
    if duration_thr_hr > duration_ten_percent:
        if duration_tmhr < duration_ten_percent:
            print("Yellow")
        else:
            print("Red")
    else:
        print("Green")
```

```python
person_no = int(input("Enter person id(0-95) to test whether reduction of training intensity is required or not: "))
training_intensity_reduction(person_no)
# for i in range(0, 96):
#   training_intensity_reduction(i)
print("Red: Training intensity reduction is required")
print("Yellow: Training intensity reduction is recommended")
print("Green: Training intensity reduction is not required")
```
